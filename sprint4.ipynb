most & least frequent MPAA Rating in the dataset in terms of occurrences

```python
df_11=df_with_year.drop(['Summary'], axis=1)

# https://stackoverflow.com/questions/49966675/pandas-least-frequent-value-in-column

test_11=df_11['MPAA Rating'].value_counts()
print(test_11)

print('most frequent MPAA Rating is',test_11.index[0],'with',test_11.iloc[0],'occurences')
print('least frequent MPAA Rating is',test_11.index[-1],'with',test_11.iloc[-1],'occurences')
```

most & least expensive Genre (take an average of all Budget measures grouped by Genre - use the groupBy() method)

```python
#Groupby genre with budget mean (by genre)
mean_by_genre=df_11.groupby('Genre')['Budget'].mean()

# find most expensive (=the highest budget) genre name and value
most_expensive_genre_amount=mean_by_genre.max()
most_expensive_genre_name=mean_by_genre.idxmax()

# find least expensive (=the lowwest budget) genre name and value
least_expensive_genre_amount=mean_by_genre.min()
least_expensive_genre_name=mean_by_genre.idxmin()

print("Most expensive genre is :",most_expensive_genre_name,"with a budget of",most_expensive_genre_amount,"USD")
print("Least expensive genre is :",least_expensive_genre_name,"with a budget of",least_expensive_genre_amount,"USD")
```

which Genre is favoured the most?

```python
# Groupby genre with rating mean (by genre)
rating_mean_by_genre=df_11.groupby('Genre')['Rating'].mean()

# find highest average rating by genre
most_favoured_genre_amount=rating_mean_by_genre.max()
most_favoured_genre_name=rating_mean_by_genre.idxmax()

print("the most favoured genre is:",most_favoured_genre_name,", with an average rating of",most_favoured_genre_amount)
# rating_mean_by_genre
```
