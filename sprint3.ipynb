Exercise 1
There are 5 numbers in the data set: (8, 12, 16, 24, 4). What is the standard deviation? And, what will be the sum of deviations of individual data points from their mean?

````python
# no NumPy used yet

e1_dataset=(8, 12, 16, 24, 4)

e1_dataset_sum=sum(e1_dataset)

e1_dataset_count=len(e1_dataset)

e1_dataset_mean=e1_dataset_sum / e1_dataset_count

e1_dataset_val_minus_mean_squared=[]

for x in e1_dataset:
    e1_dataset_val_minus_mean_squared_items = (x-e1_dataset_mean)**2
    e1_dataset_val_minus_mean_squared.append(e1_dataset_val_minus_mean_squared_items)

e1_dataset_val_minus_mean_squared_sum=sum(e1_dataset_val_minus_mean_squared)

e1_variance=e1_dataset_val_minus_mean_squared_sum / (e1_dataset_count-1)

# https://realpython.com/python-square-root-function/
# you can calculate the square root without importing the math module by using the exponent operator ** or the built-in pow() function. 
# For example, 4**0.5 will return 2.0, giving you the same result as math.sqrt()

e1_standard_deviation=e1_variance**0.5
print("What is the standard deviation?",e1_standard_deviation)

e1_dataset_datapoint_minus_mean=[]
for x in e1_dataset:
    e1_dataset_datapoint_minus_mean_item = (x-e1_dataset_mean)
    e1_dataset_datapoint_minus_mean.append(e1_dataset_datapoint_minus_mean_item)

e1_sum_of_datapoint_minus_mean=int(sum(e1_dataset_datapoint_minus_mean))
print("what will be the sum of deviations of individual data points from their mean?",e1_sum_of_datapoint_minus_mean,'\n',"for each datapoint minus mean:",list(e1_dataset_datapoint_minus_mean))
````

A) Mean
B) Median
C) Mode
D) All of these

Hint: try to create a Numpy array and calculate mean median and mode, then alter it and see what changes.

````python
import numpy as np

e8_arr = np.array([1, 2, 3, 4, 5, 2, 2, 2, 2])
print("Array:", e8_arr)

e8_sum= np.sum(e8_arr)
print("Sum of array:", e8_sum)

e8_mean= np.mean(e8_arr)
print("Mean of array:", e8_mean)

e8_median= np.median(e8_arr)
print("Median of array:", e8_median)

# no mode method found in NumPy package
#  if a single value is altered in the set, it would also alter the mode; 
#       except if the single value changed is not the most (or the second most) frequent value in the set 

print("Answer: D) All of these")
````

Exercise 9
The chart shows hourly consultancy rate of 10 people.

Calculate the standard deviation of the salaries of the 10 employees.

stat 5



````python

e9_hourly_salary_employee_count=np.array([[25,40,35,50],[3,2,1,4]])
print("Array:", e9_hourly_salary_employee_count)

e9_hourly_salary=e9_hourly_salary_employee_count[:1,:].reshape(-1)

e9_employee_count=e9_hourly_salary_employee_count[1:,:].reshape(-1)

# https://stackoverflow.com/questions/53807022/repeat-indices-with-different-repeat-values-in-numpy
e9_array_distribution_of_hourly_salary = np.repeat(e9_hourly_salary,e9_employee_count)
print("New array",e9_array_distribution_of_hourly_salary)

e9_standard_deviation_salary_employee_count=np.std(e9_array_distribution_of_hourly_salary)
print("Standard deviation of hourly rate of 10 employees:", e9_standard_deviation_salary_employee_count)
````
